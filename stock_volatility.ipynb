# **Stock Volatility Prediction System**

## Overview

The stock market is a lot like any other market: people buy and sell goods. The prices of those goods can go up or down depending on factors like supply and demand. In the case of a stock market, the goods being sold are stocks (also called equities or securities), which represent an ownership stake in a corporation.

This is an end-to-end system for predicting stock market volatility by combining data engineering, time-series modeling, and deployment.

Stock data is retrieved from the AlphaVantage API and stored in a SQLite database for persistence and efficient access. A GARCH (Generalized Autoregressive Conditional Heteroskedasticity) model is then applied to capture volatility clustering and generate predictions.

To make the solution production-ready, the model is deployed through a REST API built with FastAPI and Uvicorn, which provides endpoints for fetching data, training the model, and serving predictions.

The project is structured into four modular files: `data.py` for data extraction and storage, `model.py` for training and prediction logic, `config.py` for configuration management, and `main.py` for the FastAPI application. This notebook will start from scratch, building all functions and classes used in these modules.

**Daily Stock**

Pull daily stock price data from the AlphaVantage API using `requests`. Let's you a different stock **NVDA** - Nvidia

NVIDIA is a tech company that designs chips and software, particularly graphics processing units (GPUs), which are used for gaming, professional visualization, and powering artificial intelligence (AI) and high-performance computing.

```python
# Create URL
url = ('https://www.alphavantage.co/query?'\
'function=TIME_SERIES_DAILY&'\
f'symbol=NVDA&'\
f'outputsize=full&'\
'datatype=json&'\
f'apikey={settings.alpha_api_key}')

# Send request to API
response = requests.get(url=url)

# Extract JSON data from reponse
response_data = response.json()

# Extract daily stock data
nv_stock_data = response_data['Time Series (Daily)']

# Pull one recode of any date
nv_stock_data['2025-09-26']
```

During each trading day, the price of a stock will change, so when we're looking at whether a stock might be a good investment, we look at four types of numbers: open, high, low, close, volume.

**Open** is exactly what it sounds like: the selling price of a share when the market opens for the day. Similarly, **close** is the selling price of a share when the market closes at the end of the day, and **high** and **low** are the respective maximum and minimum prices of a share over the course of the day. **Volume** is the number of shares of a given stock that have been bought and sold that day. Generally speaking, a firm whose shares have seen a high volume of trading will see more price variation of the course of the day than a firm whose shares have been more lightly traded.

### **Distribution of stock**

**Volatility** is the measure of spread of these returns around the mean. In other words, volatility in finance is the same thing as standard deviation in statistics.

Use Colgate stock - CL, **less volatily compared to Nvidia**, to visualize distribution of the daily return.

**Create a histogram of daily returns of Colgate with 25 bins.**

```python
# Create histogram for Colgate returns
y_col = df_col['return']

plt.hist(y_col, bins=25)

plt.xlabel('Daily Return')
plt.ylabel('Frequency [count]')
plt.title('Distribution of Colgate Daily Return');
```

Looks like the daily returns follow an almost normal distribution, centered on **0**. The main difference between volatility and standard derviation is that, volatility is measured **over time**.

Now, measure the **Daily volatility** of Colgate and since the data frequency is daily, this will be exactly the same as calculating standard deviation.

As for the **Anuual volatility**, the number of trading days in a year on average is 252 (excluding the weekends)

```python
# Calculate daily volatility
col_daily_volatility = y_col.std()

# Calculate annual volatility - Number of trading days (252)
col_annual_volatility = col_daily_volatility * np.sqrt(252)

print('Colgate Daily Volatility: ', col_daily_volatility)
print('Colgate Annual Volatility: ', col_annual_volatility)
```

**Compare with the volatility of Nvidia**

```python
y_nv = df_nv['return']

# Calculate daily volatility
nv_daily_volatility = y_nv.std()

# Calculate annual volatility - Number of trading days (252)
nv_annual_volatility = nv_daily_volatility * np.sqrt(252)

print('Nvidia Daily Volatility: ', nv_daily_volatility)
print('Nvidia Annual Volatility: ', nv_annual_volatility)
```

Looks like Nvidia is more volatile than Colgate. This reinforces what we saw in the time series plot, where Nvidia returns have a much wider spread.

Since it's a time series data, another way to look at volatility is by calculating it using a rolling window. Here, we'll start focusing on Colgate exclusively.

**Calculate volatility using 50 day window**

```python
# Calculate rolling window
volatility_50d = y_col.rolling(window=50).std().dropna()

# Plot the daily returns and 50 day returns
fig, ax = plt.subplots(figsize=(15,6))
y_col.plot(ax=ax, label='Daily Return', color='C0')
volatility_50d.plot(ax=ax, label='50 Day Returns', color='C1', linewidth=3)

plt.xlabel('Date')
plt.ylabel('Returns')
plt.legend();
```

Here we can see that volatility goes up when the returns change drastically â€” either up or down.

But this plot reveals a problem. We want to use returns to see if high volatility on one day is associated with high volatility on the following day. But high volatility is caused by large changes in returns, which can be either positive or negative. How can we assess negative and positive numbers together without them canceling each other out? One solution is to take the absolute value of the numbers, which is what we do to calculate performance metrics like mean absolute error. The other solution, which is more common in this context, is to square all the values.

**Create a time series plot of the squared returns**

```python
# Plot squared returns
fig, ax = plt.subplots(figsize=(15,6))
(y_col**2).plot(ax=ax)

plt.xlabel('Date')
plt.ylabel('Squared Returns');
```

Now this is much better. Periods of high and low volatility are easily visible, PLUS high volatility days tend to cluster together which is a perfect situation to used a **GARCH** model.

### **Evaluation**

To evaluate the model we'll do walk forward validation.

**One day forecast**

```python
# Last day in the training set
y_col_train.tail(1)
```

```python
# Forecast of 1 day - variance
one_day_forecast = model.forecast(horizon=1, reindex=False).variance

one_day_forecast
```

There are two things we need to keep in mind here. First, the `model` forecast shows the predicted **variance**, not the **standard deviation** / **volatility**. So we'll need to take the square root of the value. Second, the prediction is in the form of a DataFrame. It has a DatetimeIndex, and the date is the last day for which we have training data. The `"h.1"` column stands for "horizon 1", that is, our model's prediction for the following day. We'll have to keep all this in mind when we reformat this prediction to serve to the end user of our application.

**Walk forward validation**

```python
# Create empty list to hold predictions
predictions = []

# Calculate size of test data (20%)
test_size = int(len(y_col)*0.2)

# Walk forward
for i in range (test_size):
    # Create test data
    y_train = y_col.iloc[:-(test_size - i)]

    #Train model
    model = arch_model(y_train, p=1, q=1, rescale=False).fit(disp=0)

    # Generate next prediction (volatility, not variance)
    next_pred = model.forecast(horizon=1, reindex=False).variance.iloc[0,0] **0.5


    # Append to predictions list
    predictions.append(next_pred)

# Create Series from predictions list
y_test_wfv = pd.Series(predictions, index=y_col.tail(test_size).index)

y_test_wfv.head()
```

**Visualize the predictions**

```python
fig, ax = plt.subplots(figsize=(15,6))

# plot training
y_col.tail(test_size).plot(ax=ax, label='Colgate Returns')

# plot volatility predictions * 2 std
(2* y_test_wfv).plot(ax=ax, label='2 SD Conditional volatility', color='C1', linewidth=2)

# plot  volatility predictions * -2 std
(-2* y_test_wfv.rename('')).plot(ax=ax, color='C1', linewidth=2)


plt.xlabel('Date')
plt.legend();
```

This looks pretty good. The volatility predictions seem to follow the changes in returns over time.

### **Clean Prediction**

The first thing we need to do is create a DatetimeIndex for the predictions. Using labels like `"h.1"`, `"h.2"`, etc., won't work. But there are two things we need to keep in mind. First, we can't include dates that are weekends because no trading happens on those days. And we'll need to write our dates using strings that follow the [ISO 8601](https://en.wikipedia.org/wiki/ISO_8601) standard.

**Create 5-day forecast from the model, dates written in ISO 8601 format**

```python
# Create 5-day volatility forecast
prediction = model.forecast(horizon=5, reindex=False).variance ** 0.5

# Create forecast start date
start = prediction.index[0] + pd.DateOffset(days=1)

# Create date range
prediction_dates = pd.bdate_range(start=start, periods=prediction.shape[1])

# Create prediction index label, ISO 8601 format
prediction_index = [d.isoformat() for d in prediction_dates]

prediction_index[:3]
```

Now that we know how to create the index, let's create a function to combine the index and predictions, and then return a dictionary where each key is a date and each value is a predicted volatility.

```python
def clean_prediction(prediction):

    '''Reformate model prediction to JSON.

    Parameters
    -----------
    prediction : pd.DataFrame
        Variance from a `ARCHModelForecast`
    
    Return
    -----------
    dict
        Forecast of volatility. Each key is a date in ISO 8601 format.
        Each value is predicted volatility.
    '''

    # Create forecast start date
    start = prediction.index[0] + pd.DateOffset(days=1)

    # Create date range
    prediction_dates = pd.bdate_range(start=start, periods=prediction.shape[1])

    # Create prediction index label, ISO 8601 format
    prediction_index = [d.isoformat() for d in prediction_dates]

    # Extract prediction from DataFrame, get square root
    data = prediction.values.flatten()**0.5

    # Combine `data` and `prediction_index` into series
    prediction_formatted  = pd.Series(data, index=prediction_index)

    # Return Series as dictionary
    return prediction_formatted.to_dict()
```

#### **"/fit" path**
The first path will allow the user to fit a model to stock data when they make a `post` request to the server. They'll have the choice to use new data from AlphaVantage, or older data that's already in the database. When a user makes a request, they'll receive a response telling them if the operation was successful or whether there was an error. 

One thing that's very important when building an API is making sure the user passes the correct parameters into the app. Otherwise, our app could crash! FastAPI works well with the [pydantic library](https://pydantic-docs.helpmanual.io/), which checks that each request has the correct parameters and data types. It does this by using special data classes that we need to define. The `"/fit"` path will take user input and then output a response, so we need two classes: one for input and one for output.

In the `main` model, FitIn and a FitOut data class are defined. The FitIn class inherits from the pydantic BaseClass and the FitOut class inherits from the FitIn class.

![](images/fit.png)

**Make POST request to `fit` path**

Create a post request to hit the "/fit" path running at "http://localhost:8008". You should train a GARCH(1,1) model on 2000 observations of the Apple stock data. Pass in your parameters as a dictionary using the json argument.

```python
# URL of `/fit` path
url = 'http://localhost:8008/fit'

# Data to send to path
json = {
    'ticker': 'AAPL',
    'use_new_data': True,
    'n_observations': 2000,
    'p':1,
    'q':1
}

# Response of post request
response = requests.post(url=url, json=json)

# Inspect response
print('response code: ', response.status_code)
response.json()
```
